```python
import os
import librosa
import matplotlib.pyplot as plt

# Define folder path
folder_path = r"C:\Users\user\truth-verification-project\data\MLEndDD_stories_small"

# Collect durations
durations = []
file_names = []

for filename in os.listdir(folder_path):
    if filename.endswith(".wav"):
        file_path = os.path.join(folder_path, filename)
        try:
            y, sr = librosa.load(file_path, sr=None)
            duration = len(y) / sr
            durations.append(duration)
            file_names.append(filename)
        except Exception as e:
            print(f"Error with {filename}: {e}")

# Print summary
print(f"\nChecked {len(durations)} audio files.\n")
print(f"Min duration:  {min(durations):.2f} sec")
print(f"Max duration:  {max(durations):.2f} sec")
print(f"Mean duration: {sum(durations)/len(durations):.2f} sec")

# Optional: Show histogram
plt.hist(durations, bins=20, color='skyblue', edgecolor='black')
plt.title("Distribution of Audio Durations")
plt.xlabel("Duration (seconds)")
plt.ylabel("Number of Files")
plt.grid(True)
plt.show()
```
